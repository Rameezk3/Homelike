```python
import pandas as pd
```

```python
data=pd.read_csv('task_data.csv')
data['datetime'] =pd.to_datetime(data.datetime)
data
```

```python
#filter to get only enviornment 
data= data[data['env'].isin(['production'])]
data
```

```python
# filter not internal traffic (is_internal_ip= False)

data=data[data['is_internal_ip']==False]
data
```

```python
# third filter rows without a session_id, page_type, event_type
data= data[data['session_id'].notnull() & data['page_type'].notnull() & data['event_type'].notnull()] 
data
```

```python
# Show user count by type device_class
user_device_class = pd.DataFrame(data.groupby('device_class'))
user_device_class[0]

```

```python
#  Show user count by each user's first (min datetime) non-null om_source . If the user
#  journey contains no data for om_source relabel the null value as “ no_om_source ”

user_with_dateTime = pd.DataFrame(data.groupby('device_class')['datetime'].min())


with_no_om_source=data[data['datetime'].isin(user_with_dateTime['datetime'])]
with_no_om_source['om_source']= with_no_om_source.om_source.fillna('no_om_source')
with_no_om_source
```

```python
# Show how many user journeys started on page_type=search_page and any point
#after in their user journey visited page_type=apartment_view

grouped_by_session=pd.DataFrame(data.groupby('session_id'))
grouped_by_session
```

```python
# Show how many users had a session duration of <= 5 seconds

session_dict=dict()
session_searchpage=dict()
session_apparnment_view=dict()
for entity in (grouped_by_session[1].tolist()):
    sessions=entity.sort_values(by='datetime',ascending=False)
    if len(sessions) > 0:
        session_dateTime = sessions.datetime.tolist()
        #for task a
        if (sessions.page_type.tolist()[0] == "search_page"):
            session_searchpage[sessions.session_id.any()]=True
        # for task b
        if (sessions.page_type.tolist()[-1] == "apartment_view"):
            session_apparnment_view[sessions.session_id.any()]=True
            
        count=0
        for y in range(len(session_dateTime)-1):
            diff = session_dateTime[y] - session_dateTime[y+1]
            
            if (diff.total_seconds() <= 5):
                count+=1
        session_dict[sessions.session_id.any()]=count
    else:
        pass
```

```python
session_with_count= pd.DataFrame.from_dict(list(session_dict.items())  )
session_with_count = session_with_count[session_with_count[1] >0]
session_with_count
```

```python
#How many of those user journeys started on page_type=search_page
task3a=pd.DataFrame.from_dict(list(session_apparnment_view.items())  )
task3a
```

```python
#How many of those user journeys ended on page_type=apartment_view
task3_b=pd.DataFrame.from_dict(list(session_searchpage.items())  )
```

```python
#Show how many user journeys started on page_type=search_page and any point
#after in their user journey visited page_type=apartment_view

user_searchpage=dict()
for entity in (grouped_by_session[1].tolist()):
    sessions=entity.sort_values(by='datetime',ascending=False)
    if (sessions.page_type.tolist()[0] == "search_page" and "apartment_view" in sessions.page_type.tolist()[1:] ):
        user_searchpage[sessions.session_id.any()]=True
    else:
        pass
```

```python
# user journeys started on page_type=search_page and any point after in their user 
#journey visited page_type=apartment_view
task4=pd.DataFrame.from_dict(list(user_searchpage.items())  )
task4
```

```python
 #Show how many users at any point in their user journey executed an event_type =
# ‘click_book-now’

event_clicknow=dict()
for entity in (grouped_by_session[1].tolist()):
    sessions=entity.sort_values(by='datetime',ascending=False)
    if ("click_book-now" in sessions.event_type.tolist()):
        event_clicknow[sessions.session_id.any()]=True
    else:
        pass
    
task5=pd.DataFrame.from_dict(list(event_clicknow.items())  )
task5
```

```python
#Show how many users who started on page_type = search_page executed an
#event_type = ‘request_draft-created’
user_with_searchpyge_event_type=dict()
for entity in (grouped_by_session[1].tolist()):
    sessions=entity.sort_values(by='datetime',ascending=False)
    if (sessions.page_type.tolist()[0]=='search_page' and 'request_draft-created' in sessions.event_type.tolist()):
        user_with_searchpyge_event_type[sessions.session_id.any()]=True
    else:
        pass
    
task6=pd.DataFrame.from_dict(list(user_with_searchpyge_event_type.items())  )
task6
```

## This following analysis measures your ability to extract data from a field with a JSON structure.
## You will make use of the params field data for the following analysis along with the other fields.


```python
import json

#Show how many users set a price filter above 1200 EUR and how many set a price filter
#below 1200 EUR (e.g {"event_content":{"price":1200, “currency":"EUR”}}). If the
#currency is not EUR you can ignore that session.

filtered_json_data= dict()
for entity in (grouped_by_session[1].tolist()):
    for para in entity.params:
        json_data=json.loads(para)
        if 'price' in json_data['page'] :
            if json_data['page']['price'] == None:
                pass
            elif (int(json_data['page']['price']) >1200 and json_data['page']['currency'] == 'EUR'):
                filtered_json_data[entity.session_id.any()]=True
            else:
                pass
        else:
            pass
        

```

```python
task7=pd.DataFrame.from_dict(list(filtered_json_data.items())  )
task7
```

```python
contain_req_id= dict()
for entity in (grouped_by_session[1].tolist()):
    for para in entity.params:
        json_data=json.loads(para)
        if 'request_id' in json_data['page'] :
            contain_req_id[entity.session_id.any()]=True
        else:
            pass
        

```

```python
task8=pd.DataFrame.from_dict(list(contain_req_id.items())  )
task8
```
